```python
#task10:sorting the data in the descending order based on revenue and filter the top 100 movies
```

```python
movie=df[['title','revenue','budget','runtime']].sort_values(by="revenue",ascending=False).head(500)
movie
```

```python
cet=pd.DataFrame({'Central Tendency':['Mean','Median','Mode']})
ce=cet.set_index('Central Tendency')
cet
```

```python
rev_cet=[]
rev_cet.append(movie['revenue'].mean())
rev_cet.append(movie['revenue'].median())
rev_cet.append(movie['revenue'].mode()[0])
rev_cet
```

```python
bud_cet=[]
bud_cet.append(movie['budget'].mean())
bud_cet.append(movie['budget'].median())
bud_cet.append(movie['budget'].mode()[0])
bud_cet
```

```python
run_cet=[]
run_cet.append(movie['runtime'].mean())
run_cet.append(movie['runtime'].median())
run_cet.append(movie['runtime'].mode()[0])
run_cet
```

```python
cet['Revenue_cet']=rev_cet
cet['Budget_cet']=bud_cet
cet['Runtime_cet']=run_cet
```

```python
cet
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(131)
sb.boxplot(cet['Revenue_cet'])
plt.title('Box Plot for Revenue_cet')

plt.subplot(132)
sb.boxplot(cet['Budget_cet'])
plt.title('Box Plot for Budget_cet')

plt.subplot(133)
sb.boxplot(cet['Runtime_cet'])
plt.title('Box Plot for Runtime_cet')

plt.tight_layout()
plt.show()
```
